````python
# Función con return — devuelve un resultado al código que la llamó
# Sin return, la función devuelve None (nada)

def separar_extension(nombre_archivo):
    """docstrings:
    Separa el nombre de archivo de su extensión.
    Retorna una tupla (nombre_base, extension).
    La extensión siempre sale en minúsculas.
    Si no tiene extensión, asume .mp3.
    """
    if "." not in nombre_archivo:
        return [nombre_archivo,".mp3"] # retorna dos valores a la vez
    l_nombres_ext = nombre_archivo.rsplit(".", 1)
    return [l_nombres_ext[0], "."+  l_nombres_ext[1]]


# Usar el return
nombre, ext = separar_extension("BARRIO ADENTRO 06-06-25.mp3")
print(f"nombre: '{nombre}'")
print(f"ext   : '{ext}'")
print()

# El return termina la función inmediatamente
nombre = separar_extension("audio_sin_extension")
print(f"nombre: '{nombre[0]}'")
print(f"ext   : '{nombre[1]}'")
````

````python
# Parámetros con valor por defecto
# Si no se pasa el argumento, usa el valor por defecto

def separar_extension_v2(nombre_archivo, extension_default=".mp3"):
    """Versión con extensión por defecto configurable.
    Retorna el nombre del archivo con su extensión como una sola cadena.
    """
    if "." not in nombre_archivo:
        return nombre_archivo, extension_default # Combina nombre y extensión
    nombre_base, ext = nombre_archivo.rsplit(".", 1)
    return nombre_base + "." + ext.lower() # Combina nombre base y extensión

# Sin especificar extension_default → usa '.mp3'
n_e_default = separar_extension_v2("audio_sin_ext")
print(f"default: '{n_e_default}'")

# Especificando extension_default
n_e_custom = separar_extension_v2("audio_con_ext_pordefecto", ".wav")
print(f"custom : '{n_e_custom}'")
````

## Parte 2: Diccionarios

Un **diccionario** es una colección de elementos pares `clave: valor`.
Se define con llaves `{}`.

Es como una tabla de traducción: le das una clave y te devuelve
el valor correspondiente, instantáneamente, sin recorrer nada.

```python
{"clave": "valor",  "agosto": "08",  "barrio adentro": "barrio_adentro"}
```

**Diferencia clave con las listas:**
- Lista: accedo por posición numérica ->`lista[0]`
- Diccionario: accedo por clave → `diccionario["agosto"]`

````python
dict_ex = {
    "nombre": "Carolina",
    "Edad": 36,
    "documento": 111111111
}
dict_ex
````

**Propiedad de los diccionarios:**



*   Dinámicos, se pueden añadir o eliminar elementos.
*   Indexados, son accesibles a através de las claves
*   Anidados, dentro del value se pueden añadir otros diccionarios.





**Acceder y modificar elementos**
Se puede acceder a sus elementos con `[]` o también con la función `get()`

````python
print(dict_ex['nombe'])
````

````python
print(dict_ex.get('nombre'))
````

### **Metodos de los dicconarios**
`keys()`

El método `keys()` devuelve una lista con todas las claves del diccionario.

````python
diccionario = {'a': 1, 'b': 2}
k = diccionario.keys()
print(k)
print(list(k))
````

**`values()`**

El método `values()` devuelve una lista con todos los values o valores del diccionario.

````python
d = {'a': 1, 'b': 2}
print(list(d.values()))
````

**`pop(<key>[,<default>])`**

El método pop() busca y elimina la key que se pasa como parámetro y devuelve su valor asociado. Daría un error si se intenta eliminar una key que no existe.


````python
d = {'a': 1, 'b': 2}
d.pop('a')
print(d)
````

También se puede pasar un segundo parametro a la función, y el valor por defecto que le damos en caso de que la clave no exista, así se evitaria el error:


````python
d = {'a': 1, 'b': 2}
d.pop('c', -1)
print(d)
````

**`update()`**

El método update() se llama sobre un diccionario y tiene como entrada otro diccionario. Los value son actualizados y si alguna key del nuevo diccionario no esta, es añadida.

````python
d1 = {'a': 1, 'b': 2}
d2 = {'a': 0, 'd': 400}
d1.update(d2)
print(d1)
````
