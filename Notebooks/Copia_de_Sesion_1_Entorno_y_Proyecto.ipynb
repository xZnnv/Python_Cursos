### 2.2 — Variables:
Una variable es un nombre que le damos a un valor para poder usarlo después.

*Una* **variable** *es como una cajita con nombre. Guardas algo adentro y después lo puedes usar cuando quieras*.

```python

```

El símbolo `=` significa **"guarda este valor en esta variable"**. No es el "igual" de las matemáticas — es una instrucción: *"toma esto y guárdalo aquí"*.

Entendiendo las variables:

- Almacena y reutiliza la información.
- Fácil de leer y actualizar.
- Evita escribir cosas una y otra vez.

**Reglas para nombrar variables:**

| | Ejemplo |

| ✅ Letras, números y guion bajo `_` | `nombre_programa`, `conductora_1` |

#### Combinando texto y variables con **f-strings**: la forma moderna y la que vamos a usar

### 2.3 — Tipos de datos: texto, números y verdadero/falso

No todos los datos son iguales. Python distingue entre **tipos** de datos. Los más importantes para nuestro proyecto:

| Tipo | Nombre en Python | Ejemplo en Oriente Stereo |
|---|---|---|
| Texto | `str` (string) | `"CronicasOriente"`, `"Fernanda"` |
| Número entero | `int` (integer) | `30` (MB que pesa un MP3 de una hora) |
| Número decimal | `float` | `88.4` (frecuencia de la emisora) |
| Verdadero/Falso | `bool` (boolean) | `True` si tiene audio, `False` si no |

**¿Por qué importa el tipo?** Porque Python hace cosas distintas según el tipo. Por ejemplo, puedes sumar dos números, pero sumar un número con un string da error.

Puedes saber el tipo de cualquier valor con la función `type()`:


````python
# Diferentes tipos de datos en nuestro proyecto

# str — texto (siempre entre comillas)
nombre_programa = "CronicasOriente"
formato_audio   = "mp3"

# int — número entero (sin decimales, sin comillas)
tamano_mp3_mb = 30    # Un programa de 1 hora pesa ~30 MB en MP3
tamano_wav_mb = 600   # El mismo programa pesa ~600 MB en WAV

# float — número decimal (punto, no coma)
frecuencia_fm = 88.4  # Frecuencia de la emisora

# bool — verdadero o falso (sin comillas, con mayúscula inicial)
tiene_audio  = True   # ¿Este registro tiene archivo de audio?
tiene_backup = False  # ¿Este archivo tiene copia de seguridad?

# Veamos los tipos
print(type(tiene_audio))

print(f"'{nombre_programa}' es de tipo: {type(nombre_programa)}")
print(f"'{tamano_mp3_mb}' es de tipo: {type(tamano_mp3_mb)}")
print(f"'{frecuencia_fm}' es de tipo: {type(frecuencia_fm)}")
print(f"'{tiene_audio}' es de tipo: {type(tiene_audio)}")
````

````python
# Con números podemos hacer operaciones matemáticas
# Esto va a ser útil para calcular espacio en el sistema de backup

tamano_mp3_mb = 30
tamano_wav_mb = 600

diferencia    = tamano_wav_mb - tamano_mp3_mb
cuantas_veces = tamano_wav_mb / tamano_mp3_mb

print(f"Un WAV pesa {tamano_wav_mb} MB y un MP3 pesa {tamano_mp3_mb} MB")
print(f"Diferencia: {diferencia} MB")
print(f"El WAV pesa {cuantas_veces} veces más que el MP3")
print("")
print("Para Oriente Stereo esto importa mucho:")
print("su Google Drive está al 80% y los WAV son el mayor riesgo.")
````

````python
# ✏️ TU TURNO
# Oriente Stereo tiene 50 programas de una hora guardados en WAV
# y 80 programas guardados en MP3.
# Calcula cuántos GB ocupan en total.
# (1 GB = 1024 MB)

programas_wav = 50
programas_mp3 = 80
tamano_wav_mb = 600
tamano_mp3_mb = 30

# Calcula el total en MB
total_wav_mb = (programas_wav * tamano_wav_mb)/1024 # Pista: programas_wav * tamano_wav_mb
total_mp3_mb = (programas_mp3 * tamano_wav_mb)/1024  # Pista: programas_mp3 * tamano_mp3_mb
total_mb     = total_wav_mb + total_mp3_mb # Pista: total_wav_mb + total_mp3_mb

# Convierte a GB
total_gb = total_mb      # Pista: total_mb / 1024


print(f"Total WAV    : {total_wav_mb:.2f} GB")
print(f"Total MP3    : {total_mp3_mb:.2f} GB")
print(f"Total general: {total_mb:.2f} GB")
print(f"En GB        : {total_gb:.2f} GB")  # :.2f muestra solo 2 decimales
````

### 2.4 — El problema de los nombres: por qué los strings importan tanto

En el archivo de Oriente Stereo, el mismo programa puede tener archivos con nombres completamente distintos:

```
2025-11-10_Noticias.mp3
Noticias_10nov2025.mp3
NotiOriente.mp3
audio_grabacion_lunes.mp3
```

Todos podrían ser el mismo programa, pero es imposible saberlo mirando el nombre. **Esto hace imposible organizar, buscar o hacer backup de forma automática.**

En la Sesión 2 vamos a construir una función que convierta cualquier nombre al formato estándar:
```
PROGRAMA_AAAA-MM-DD_CONDUCTOR.mp3
→ CronicasOriente_2025-11-10_Fernanda.mp3
```

Hoy aprendemos las operaciones básicas con strings que vamos a necesitar para eso:


**Los strings tienen métodos**:

> "Un método es una función que viene incorporada en el string. Se usa con un punto después del string."



````python
#Ejemplo de metodos de string:
nombre_programa =  " CRONICAS De Oriente "

#imprimir el nombre del programa en minuscula:
print(nombre_programa.title())
#Ejemlos sobre los diferentes tipos de metodos:
print(nombre_programa.upper())

#imprimir el nombre del programa en mayuscula:
print(nombre_programa.upper())


#imprimir el nombre del programa como titulo( las primeras letras en mayuscula)
print(nombre_programa.title())

#print(type(nombre_programa.strip()))

#reemplazar espacios por
````

````python
# Operaciones básicas con strings — las herramientas de la sesión 2

nombre_archivo = "NotiOriente.mp3"

print(f"Original       : {nombre_archivo}")

# .lower() — todo en minúsculas
print(f"En minúsculas  : {nombre_archivo.lower()}")

# .upper() — todo en mayúsculas
print(f"En mayuscula : {nombre_archivo.upper()}")
# len() — cuenta los caracteres
print(f"Largo          : {len(nombre_archivo)} caracteres")

# .replace() — reemplaza texto dentro del string. Reemplazar Noti por Noticias
print(f"Sin Noti       : {nombre_archivo.replace('Noti', 'Noticias')}")
print("---")

# .split() — divide el string en una lista de partes
nombre_con_guiones = "CronicasOriente_2025-11-10_Fernanda.mp3"
partes = nombre_con_guiones.split("_")
print(f"Nombre         : {nombre_con_guiones}")
print(f"Dividido por _ : {partes}")
print(f"Parte 0        : {partes[0]}")   # El primero es índice 0
print(f"Parte 1        : {partes[1]}")
print(f"Parte 2        : {partes[2]}")
````

> 💡 **¿Por qué `partes[0]` y no `partes[1]`?**
>
> En Python, las listas empiezan a contar desde **0**, no desde 1.
> Si la lista es `["CronicasOriente", "2025-11-10", "Fernanda.mp3"]`:
> - `partes[0]` → `"CronicasOriente"`
> - `partes[1]` → `"2025-11-10"`
> - `partes[2]` → `"Fernanda.mp3"`
>
> Esto lo vamos a usar mucho a partir de la sesión 2.


````python
# ✏️ TU TURNO
# Trabaja con estos nombres de archivos reales de la emisora

archivo_1 = "CronicasOriente_2025-03-15_Fernanda.mp3"
archivo_2 = "NOTIORIENTE.WAV"
archivo_3 = "programa radio comunitaria sabado.mp3"

# 1. Muestra cada archivo en minúsculas
print("--- En minúsculas ---")
print(archivo_1.lower())

# Haz lo mismo para archivo_2 y archivo_3
print(archivo_2.lower())
print(archivo_3.lower())


print("--- Longitudes ---")
# 2. ¿Cuántos caracteres tiene cada nombre?
print(f"Longitud de archivo_2: {len(archivo_2)}")
# Haz lo mismo para archivo_2 y archivo_3
print(f"longitud de archivo_1: {len(archivo_1)}")
print(f"longitud de archivo_3: {len(archivo_3)}")


print("--- División ---")
# 3. Divide archivo_1 por "_" y muestra cada parte
partes = archivo_2.split("_")
kamu = archivo_1.split("_")
print(f"partes: {kamu}")
print(f"nombre del programa: {kamu[0]}")
print(type(kamu))
print(f"Partes: {partes}")
print(f"Nombre del programa: {partes[0]}")
# ¿Cuál es la fecha? ¿Y la conductora?

````

### ¿Qué viene en la Sesión 2?

Vamos a construir la función `normalizar_nombre()` — la primera pieza real del sistema.

Esta función va a tomar cualquier nombre de archivo en cualquier formato y lo va a convertir al estándar:
```
"NotiOriente.mp3"  →  "NoticiasOriente_desconocido_desconocido.mp3"
"CronicasOriente_2025-11-10_Fernanda.mp3"  →  ya está en formato estándar ✓
```

Para eso vamos a necesitar todo lo que aprendiste hoy, más algo nuevo: **funciones** y **condicionales** (`if`/`else`).
