**Al final de esta sesión 3 habremos construido `normalizar_nombre()` — la función que convierte cualquier nombre al formato estándar:**

```
nombre_programa_AAAA-MM-DD_descripcion.ext
nombre_programa_AAAA-MM-DD_epNNN_descripcion.ext  ← si tiene episodio
nombre_programa_sin_fecha_descripcion.ext          ← si no hay fecha
```

**Reglas del estándar:**
- Todo en **minúsculas**
- Palabras separadas por **guión bajo** `_`
- Fecha en formato **AAAA-MM-DD** — se ordena sola en cualquier explorador
- Sin fecha → **`sin_fecha`** (nunca se inventa)
- Con episodio → **`ep001`, `ep122`, `ep293`** (siempre 3 dígitos)
- Sin tildes, sin puntos intermedios, sin comas, sin paréntesis

> ⚠️ **Decisión de diseño importante:** el nombre del archivo NO puede decirle al sistema a qué programa pertenece por sí solo. `EPISODIO 34-2025.mp3` podría ser de *A Prender La Onda* o de cualquier otro programa. ## Parte 1 — Métodos de string

```python
texto.metodo()
```

Métodos de String que vamos a usar en nuestro proyecto:

1.   **`.lower()`**-> Todo en minúscula
2.  **`.upper()`** - > Todo en mayúscula
3.  .**`strip()`** ->  Quita caracteres al inicio y al final de la cadena.
4.  **`.replace("viejo", "nuevo")`** -> reemplazar el texto dentro de la cadena.
5.  **`.split(separador)`** -> partir la cadena en piezas usando un separador. Retorna una lista con las piezas.
6.  **`.startswith()`, `.endswith()`, `.in`** -> Para detectar patrones dentro del nombre del archivo. Devuelve verdadero o falso. Verifica si la cadena empieza o termina con una subcadena.
7.  **`.rstrip()`** -> Elimna los espacio en blanco que hayan al final de la cadena.
8.  **`.isdigit()`** -> Pregunta si todos los caracteres de la cadena son digitos?. ¿ Para qué podría ser útil para nuestro problema de la emisora?
9.  **`.rsplit(".",1)`** -> Dividir la cadena una sóla vez desde la derecha. ¿Para que seria útil?
10.  **`.zfill(width)`** ->  rellenar una cadena de con ceros iniciales hasta alcanzar el ancho total especificado, Resulta útil para formatear números (identificadores y fechas).

## Parte 2 — Listas

Una **lista** es una colección ordenada de elementos entre corchetes `[]`.
Puede contener strings, números, booleans — cualquier tipo.

Cada elemento tiene una **posición** (índice) que empieza en **0**, no en 1:

```
  ['.mp3',  '.wav',  '.ogg',  '.flac']
      0        1        2        3
     -4       -3       -2       -1   ← índices negativos
```

Las listas son mutables, es decir, se pueden añadir elementos nuevos a la lista, eliminar elementos, reemplazar, cambiar de posición.
Las listas soportan una variedad de métodos incorporados en Python que llevan ciertas operaciones sobre ellas. Esto implica que la longitud de una lista puede cambiar.


### Las listas en Python soportan desempaquetado.
Esencialmente, cuando se crea una lista, varios valores son 'empaquetados' en una única variable, como ya he mencionado anteriormente.
Esos valores puede ser 'desempaquetados' y asignados a variables individuales.


```python
arq_archivo= ["en_la_raya", "oriente_stereo", "08-02-2025"]

programa,conductor,fecha = arq_archivo

```










````python
arq_archivo = ["en_la_raya", "oriente_stereo", "08-02-2025"]
programa,conductor,fecha = arq_archivo
print(f"programa: {programa}\nconductor: {conductor}\nfecha: {fecha}")
````

### Actualizar listas en Python
Ya que las listas son mutables vamos a conocer algunas formas básicas de actualziar los datos contenidos en una lista.

Para actualizar un elemento en particular, hacemos referencia a su índice entre corchetes y asignamos el nuevo valor.


> nombre_lista[indice] = nuevo_valor



Métodos de una lista:


*   `.append()` -> añade un elemento nuevo al final de la lista.
      ```
      #sintaxis general
      nombre_lista.append(elemento)
      ```
    
*   `.insert()` -> añadir un elemento en una posición determinada
      ```
      # sintaxis general
      nombre_lista.insert(index,item)
      ```


*   `.remove()` -> elimina la primera instancia del elemnto que se le proporciona como argumento.
      ```
      # sintaxis general:
      nombre_lista.remove(elemnento)
      ```




````python
# Crear una lista y acceder a sus elementos

extensiones_audio = [".mp3", ".wav", ".ogg", ".wma", ".flac", ".mp4"]

print(extensiones_audio)           # lista completa
print(len(extensiones_audio))      # cuántos elementos tiene → 6
print()

# Acceder por índice
print(extensiones_audio[0])        # '.mp3'  → el primero (índice 0)
print(extensiones_audio[1])        # '.wav'
print(extensiones_audio[-1])       # '.mp4'  → el último
print(extensiones_audio[-2])       # '.flac' → el penúltimo
print()

# 'in' → pregunta si un elemento está en la lista
print(".mp3" in extensiones_audio)  # True
print(".jpg" in extensiones_audio)  # False
print(".MP3" in extensiones_audio)  # False ← mayúsculas importan
````

## Parte 3 — Condicionales: `if` / `elif` / `else`

Los condicionales le dicen a Python: *«si esta condición es verdadera,
ejecuta este bloque; si no, prueba la siguiente»*.

```python
if condicion_1:
    # se ejecuta si condicion_1 es True
elif condicion_2:
    # se ejecuta si condicion_1 fue False pero condicion_2 es True
else:
    # se ejecuta si ninguna condición anterior fue True
```

**Operadores lógicos:**
- `and` → las dos condiciones deben ser `True`
- `or` → al menos una debe ser `True`
- `not` → invierte el resultado

````python
# Condicionales con 'and', 'or', 'not'

nombre = "Episodio 122 2023 (1).mp3"
n = nombre.lower()

# ¿Tiene la palabra 'episodio' Y es audio?
if "episodio" in n and nombre.endswith(".mp3"):
    print("Es un episodio de audio")

# ¿Empieza con un número seguido de ' - '? (patrón de Diana Uribe)
nombre_diana = "7 - Comienzos de la radio en Colombia.mp3"
partes = nombre_diana.split(" - ")
if len(partes) >= 2 and partes[0].strip().isdigit():
    numero = int(partes[0].strip())
    print(f"Es episodio {numero} → ep{numero:03d}")
````

    # la indentación (4 espacios) es obligatoria

````python
# ─── RETO PARTE 1: detectar_fecha() ────────────────────────────────
#
# Necesita manejar los formatos reales del disco:
#
#   '06-06-25'              → '2025-06-06'   (Patrón A: DD-MM-AA)
#   'AGOSTO 25 DE 2025'     → '2025-08-25'   (Patrón B: MES DD DE AAAA)
#   '14 DE MARZO 2026'      → '2026-03-14'   (Patrón B: DD DE MES AAAA)
#   'Episodio 35 2025'      → '2025'         (Patrón C: solo año)
#   'mujeres bolero parte 2'→ 'sin_fecha'    (sin fecha)
#
# Regla: NUNCA inventa una fecha. Si no hay → 'sin_fecha'

# Meses en español (los necesitas para el Patrón B)
# (En S3 aprenderemos diccionarios a fondo — aquí los usamos como herramienta)
MESES = {
    "enero":"01",     "ene":"01",
    "febrero":"02",   "feb":"02",
    "marzo":"03",     "mar":"03",
    "abril":"04",     "abr":"04",
    "mayo":"05",
    "junio":"06",     "jun":"06",
    "julio":"07",     "jul":"07",
    "agosto":"08",    "ago":"08",
    "septiembre":"09","sept":"09",  "sep":"09",
    "octubre":"10",   "oct":"10",
    "noviembre":"11", "nov":"11",
    "diciembre":"12", "dic":"12",
}
# Cómo usar el diccionario:
# MESES["agosto"]       → '08'
# MESES.get("agosto")   → '08'
# MESES.get("xyz")      → None (no falla si la clave no existe)


def detectar_fecha(nombre_archivo):
    """
    Detecta la fecha en el nombre de un archivo de Oriente Estéreo.
    Retorna la fecha en formato 'AAAA-MM-DD', solo el año, o 'sin_fecha'.
    NUNCA inventa una fecha.

    Herramientas que necesitas:
        - .rsplit('.', 1)  → separar el nombre de la extensión
        - .split()         → partir en palabras
        - .split('-')      → partir por guión
        - .isdigit()       → saber si una pieza es número
        - .lower()         → comparar sin importar mayúsculas
        - .zfill(2)        → '6' → '06' (rellenar con cero a la izquierda)
        - MESES.get(p)     → convertir nombre de mes a número
    """
    # Separar extensión primero
    if "." in nombre_archivo:
        nombre_base, _ = nombre_archivo.rsplit(".", 1)
    else:
        nombre_base = nombre_archivo

    # ── Patrón A: buscar segmento DD-MM-AA en el nombre ──────────────
    # Estrategia: partir el nombre en palabras, y para cada palabra
    # verificar si tiene exactamente 2 guiones y 3 partes numéricas




    # ── Patrón B: buscar mes en español ──────────────────────────────
    # Estrategia: normalizar el texto (minúsculas, sin guiones/puntos),
    # partir en palabras, y buscar alguna que sea un mes conocido.
    # Cuando encuentres el mes, buscar el día y el año cerca.
    # Subpatrón MES DIA DE AAAA  (agosto 25 de 2025)
    # Subpatrón DD DE MES AAAA  (14 de marzo 2026)




    # ── Patrón C: solo año de 4 dígitos ──────────────────────────────
    # Si el nombre tiene un año solo (2025, 2026) sin mes ni día
    # guardamos solo el año — es lo que tenemos, en caso contrario retornar "sin_fecha"





# ── Verificación ────────────────────────────────────────────────────
print("=== Verificación detectar_fecha() ===")
casos = [
    ("BARRIO ADENTRO 06-06-25.mp3",                              "2025-06-06"),
    ("BARRIO ADENTRO 17-07-25.mp3",                              "2025-07-17"),
    ("DEJAME SER ARTE 03-07-25.mp3",                             "2025-07-03"),
    ("DEJAME SER ARTE17-07-25.mp3",                              "2025-07-17"),
    ("DSA 02-10-25.mp3",                                         "2025-10-02"),
    ("EN LA RAYA DEPORTES- AGOSTO 25 DE 2025.mp3",              "2025-08-25"),
    ("EN LA RAYA DEPORTES DIC 1 DE 2025.mp3",                   "2025-12-01"),
    ("En la Raya Deportes- Septiembre 29 de 2025.mp3",          "2025-09-29"),
    ("PROG. 293 ENTRE NOSOTROS, SABADO 14 DE MARZO 2026.mp3",   "2026-03-14"),
    ("Episodio 35 2025.mp3",                                     "2025"),
    ("Episodio 122 2023 (1).mp3",                                "2023"),
    ("PAGANO BOLERO - mujeres bolero parte 2.mp3",              "sin_fecha"),
    ("1 - La historia de la radio.mp3",                         "sin_fecha"),
    ("CABEZOTE DSA.mp3",                                        "sin_fecha"),
]
ok_total = True
for nombre, esperado in casos:
    resultado = detectar_fecha(nombre)
    ok = resultado == esperado
    if not ok: ok_total = False
    icono = "✅" if ok else "❌"
    print(f"  {icono} esp='{esperado:12s}' obt='{resultado:12s}' | {nombre}")
print()
print("🎉 Todos correctos" if ok_total else "⚠️  Revisar los ❌")
````
